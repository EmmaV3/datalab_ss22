```python
# Import data from txt-file, return one list of x and one list of y values.
def import_plot_data(file):
    input_data = np.loadtxt(file)
    X = []
    Y = []
    output = []
    
    for i in range(len(input_data)):
        for j in range(2):
            if j == 0:
                X.append(input_data[i][j])
            else:
                Y.append(input_data[i][j])
    
    output = [X,Y]
    
    return(output)
```

```python
def import_plot_data_in_range(file, data_point_range):
    input_data = np.loadtxt(file)
    X = []
    Y = []
    output = []
    
    for i in range(data_point_range[0], data_point_range[1]):
        for j in range(2):
            if j == 0:
                X.append(input_data[i][j])
            else:
                Y.append(input_data[i][j])
    
    output = [X,Y]
    
    return(output)
```

```python
# Plot the given xy data.
def custom_subplots(xlist, ylist, title_list, **plt_kwargs):
    if (len(xlist) != len(ylist)):
        print("The number of x data does not match the number of y data.")
        return
    
    fig, axs = plt.subplots(len(xlist)//2, 2,figsize = (12,16))
    
    for i in range(0, len(xlist), 2):
    
        axs[i//2, 0].plot(xlist[i], ylist[i])
        axs[i//2, 0].set_title(title_list[i])
        axs[i//2, 1].plot(xlist[i+1], ylist[i+1])
        axs[i//2, 1].set_title(title_list[i+1])
   

    #for ax in axs.flat:
        #ax.set(xlabel= x_label, ylabel= y_label)
        
    return 
```

```python
[X1,Y1] = import_plot_data_in_range("data_f50_a50_s05.txt",[0,200]);
[X2,Y2] = import_plot_data_in_range("data_f50_a50_s8.txt",[0,200]);
[X3,Y3] = import_plot_data_in_range("data_f50_a001_s01.txt",[0,200]);
[X4,Y4] = import_plot_data_in_range("data_f50_a001_s4.txt",[0,200]);
```

```python
Xl = [X1,X2,X3,X4];
Yl = [Y1,Y2,Y3,Y4];
custom_subplots(Xl, Yl, ["f = 50, a = 50, s = 0.5","f = 50, a = 50, s = 8","f = 50, a = 0.01, s = 0.1","f = 50, a = 0.01, s = 4"])
```
